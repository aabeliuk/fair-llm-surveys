```python
import pandas as pd
import numpy as np
# import seaborn as sns
```

```python
def get_reg_df_chile(df, pred_variable, pred_column_name="pred"):
  df_reg = pd.DataFrame()
  df_reg = df[['sexo', 'iden_pol_2', 'gse', 'religion_82', 'nom_region','edad','info_enc_58','esc_nivel_1','zona_u_r', pred_variable, pred_column_name]].copy()
  df_reg[pred_variable] = df_reg[pred_variable].astype(int)
  df_reg[pred_column_name] = df_reg[pred_column_name].astype(int)

  df_reg['accuracy'] = 0  # Initialize the new column with 0s
  df_reg.loc[df_reg[pred_variable] - df_reg[pred_column_name] == 0, 'accuracy'] = 1


  # Rename index in the DataFrame
  df_reg['sexo'] = df_reg['sexo'].replace({1: "Hombre", 2: "Mujer"})

  df_reg['iden_pol_2'] = df_reg['iden_pol_2'].replace({1: "Izquierda", 2: "Izquierda", 3: "Izquierda", 4: "Izquierda",
              5: "Centro", 6: "Centro",
              7: "Derecha", 8: "Derecha", 9: "Derecha", 10: "Derecha",
              88: "Ninguna ideología", 99: "Ninguna ideología"})

  df_reg['religion_82'] = df_reg['religion_82'].replace({1: "Religioso", 2: "Religioso", 3: "Religioso", 4: "Religioso", 5: "Religioso",
              6: "Religioso", 7: "Religioso", 10: "Ateo/agnóstico/SR/NC",
              11: "Ateo/agnóstico/SR/NC", 8: "Ateo/agnóstico/SR/NC",
              9: "Ateo/agnóstico/SR/NC", 88: "Ateo/agnóstico/SR/NC", 99: "Ateo/agnóstico/SR/NC"})

  df_reg['nom_region'] = df_reg['nom_region'].replace(
    {value: 'Provincia' for value in df_reg['nom_region'].unique() if value != 'METROPOLITANA'}
)
  df_reg['nom_region'] = df_reg['nom_region'].replace({ 'METROPOLITANA': 'Metropolitana'})

  df_reg['zona_u_r'] = df_reg['zona_u_r'].replace({ 1: 'Urbano', 2: 'Rural'})

  df_reg['gse'] = df_reg['gse'].replace({1: "Alto", 2: "Medio", 3: "Medio", 4: "Bajo",  5: "Bajo"})

  # Apply conditions to create the 'grupo_edad' column
  df_reg.loc[(df_reg["edad"] >= 18) & (df_reg["edad"] <= 26), "grupo_edad"] = "joven"
  df_reg.loc[(df_reg["edad"] >= 27) & (df_reg["edad"] <= 59), "grupo_edad"] = "adulto"
  df_reg.loc[(df_reg["edad"] >= 60), "grupo_edad"] = "mayor"

  df_reg = df_reg.drop(columns = [pred_variable,pred_column_name])

  df_reg = df_reg.rename(columns={"info_enc_58": "Indigenous"})
  df_reg['Indigenous'] = df_reg['Indigenous'].replace({1: "Indigenous", 2: "Non-indigenous", 3: "None",  8: "None"})

  df_reg = df_reg.rename(columns={"esc_nivel_1": "Education"})
  df_reg['Education'] = df_reg['Education'].replace({1: "low education", 2: "low education",3: "low education",0: "low education",
                                                     4: "medium education",5: "medium education",7: "medium education",
                                                     6: "high education",8: "high education",9: "high education",10: "high education",
                                                     99:"None",88:"None",-9:"None",-8:"None"})


  return df_reg


def get_reg_df_eeuu(df, pred_variable, pred_column_name="pred"):
  df_reg = pd.DataFrame()
  df_reg = df[['sexo', 'iden_pol_2', 'gse', 'religion_82', 'nom_region','edad','race','esc_nivel_1','zona_u_r', pred_variable, pred_column_name]].copy()
  df_reg[pred_variable] = df_reg[pred_variable].astype(int)
  df_reg[pred_column_name] = df_reg[pred_column_name].astype(int)

  df_reg['accuracy'] = 0  # Initialize the new column with 0s
  df_reg.loc[df_reg[pred_variable] - df_reg[pred_column_name] == 0, 'accuracy'] = 1


  # Rename index in the DataFrame
  df_reg['sexo'] = df_reg['sexo'].replace({1: "Hombre", 2: "Mujer"})

  df_reg['iden_pol_2'] = df_reg['iden_pol_2'].replace({1: "Izquierda", 2: "Izquierda", 3: "Izquierda",
              4: "Centro",
              5: "Derecha", 6: "Derecha", 7: "Derecha",
              88: "Ninguna ideología", 99: "Ninguna ideología", -8: "Ninguna ideología", -9: "Ninguna ideología"})



  df_reg['religion_82'] = df_reg['religion_82'].replace({1: "Religioso", 2: "Religioso", 3: "Religioso", 4: "Religioso", 5: "Religioso", 6: "Religioso", 7: "Religioso", 8: "Religioso",
              9: "Ateo/agnóstico/SR/NC", 10: "Ateo/agnóstico/SR/NC", 11: "Ateo/agnóstico/SR/NC", 12: "Ateo/agnóstico/SR/NC", 88: "Ateo/agnóstico/SR/NC", 99: "Ateo/agnóstico/SR/NC",
              -8: "Ateo/agnóstico/SR/NC", -9: "Ateo/agnóstico/SR/NC"})

  df_reg['nom_region'] = df_reg['nom_region'].replace(
    {value: 'Provincia' for value in df_reg['nom_region'].unique() if value != 'METROPOLITANA'}
)
  df_reg['nom_region'] = df_reg['nom_region'].replace({ 'METROPOLITANA': 'Metropolitana'})

  df_reg['zona_u_r'] = df_reg['zona_u_r'].replace({ 1: 'Urbano', 2: 'Urbano', 3: 'Rural', 4: 'Rural'})


  df_reg['gse'] = df_reg['gse'].replace({20: "Alto", 21: "Alto", 22: "Alto", 88:np.nan, 99:np.nan, -8:np.nan, -9:np.nan})
  # Replace gse values between 10 and 19 (inclusive) with "medio"
  for i in range(10, 20):
    df_reg['gse'] = df_reg['gse'].replace({i: "Medio"})
  # Replace gse values between 1 and 9 (inclusive) with "bajo"
  for i in range(1, 10):
    df_reg['gse'] = df_reg['gse'].replace({i: "Bajo"})


  # Apply conditions to create the 'grupo_edad' column
  df_reg.loc[(df_reg["edad"] >= 18) & (df_reg["edad"] <= 26), "grupo_edad"] = "joven"
  df_reg.loc[(df_reg["edad"] >= 27) & (df_reg["edad"] <= 59), "grupo_edad"] = "adulto"
  df_reg.loc[(df_reg["edad"] >= 60), "grupo_edad"] = "mayor"

  df_reg = df_reg.drop(columns = [pred_variable,pred_column_name])


  df_reg['race'] = df_reg['race'].replace({1: "White", 2: "Non-White", 3: "Non-White",4: "Non-White",5: "Non-White",6: "Non-White",
                                           -8: "None",-9: "None"})


  df_reg = df_reg.rename(columns={"esc_nivel_1": "Education"})
  df_reg['Education'] = df_reg['Education'].replace({1: "low education",
                                                     2: "medium education",3: "medium education",4: "medium education",5: "medium education",
                                                     6: "high education",7: "high education",8: "high education",
                                                     95:"None",-8:"None",-9:"None",88:"None",99:"None" })



  return df_reg
```

# Logistic regression

```python
#Read data from CSV files
ruta = "Data/Experiments/"

#Chile
llama_1_csv = pd.read_csv(ruta +"llama_presidencial.csv")
llama_2_csv = pd.read_csv(ruta +"llama_aborto.csv")
llama_3_csv = pd.read_csv(ruta +"llama_plebiscito.csv")

#USA
llama_1_eeuu_csv = pd.read_csv(ruta +"llama_presidencial_eeuu.csv")
llama_2_eeuu_csv = pd.read_csv(ruta +"llama_aborto_eeuu.csv")
```

## Chile

```python
#combine dataframes for Chile
df_reg = pd.DataFrame()
df_reg = get_reg_df_chile(llama_1_csv, 'elec_pres_144_a')
df_reg['Q'] = "presidencial"
df_reg['Country'] = "Chile"

df = get_reg_df_chile(llama_2_csv, 'religion_14')
df['Q'] = 'aborto'
df['Country'] = "Chile"
df_reg = pd.concat([df_reg, df], ignore_index=True)

df = get_reg_df_chile(llama_3_csv, 'constitucion_20_a')
df['Q'] = "plesbicito"
df['Country'] = "Chile"
df_reg = pd.concat([df_reg, df], ignore_index=True)


df_reg
```

```python
# import statsmodels.api as sm
# from patsy import dmatrices
import statsmodels.formula.api as smf


df_reg['Country'] = pd.Categorical(df_reg['Country'], categories=['EEUU','Chile'], ordered=False)
df_reg['sexo'] = pd.Categorical(df_reg['sexo'], categories=['Hombre','Mujer'], ordered=False)
df_reg['zona_u_r'] = pd.Categorical(df_reg['zona_u_r'], categories=['Urbano','Rural'], ordered=False)
df_reg['Indigenous'] = pd.Categorical(df_reg['Indigenous'], categories=['Non-indigenous','Indigenous'], ordered=False)
# df_reg['race'] = pd.Categorical(df_reg['race'], categories=['White','Non-White'], ordered=False)

df_reg['grupo_edad'] = pd.Categorical(df_reg['grupo_edad'], categories=['joven','adulto','mayor'], ordered=False)
df_reg['iden_pol_2'] = pd.Categorical(df_reg['iden_pol_2'], categories=['Ninguna ideología','Izquierda','Centro','Derecha',], ordered=False)


# mod = smf.logit(formula='accuracy ~  C(Q) + C(sexo) + C(grupo_edad) + C(zona_u_r) + C(Indigenous) + C(Education) + C(gse) + C(iden_pol_2)  + C(religion_82) -1', data=df_reg)

# res = mod.fit()

# print(res.summary())
```

```python
# Model 1: Only main effects
mod1 = smf.logit(
    formula='''
    accuracy ~ C(Q)
    + C(sexo)
    + C(grupo_edad)
    + C(zona_u_r)
    + C(Indigenous)
    + C(Education)
    + C(gse)
    + C(iden_pol_2)
    + C(religion_82)
    -1
    ''',
    data=df_reg
)

res = mod1.fit()

print(res.summary())
```

```python
# Model 2: Add interaction between sexo and religion_82
mod2 = smf.logit(
    formula='''
    accuracy ~ C(Q)
    + C(sexo)
    + C(grupo_edad)
    + C(zona_u_r)
    + C(Indigenous)
    + C(Education)
    + C(gse)
    + C(iden_pol_2)
    + C(religion_82)
    + C(sexo):C(religion_82)
    -1
    ''',
    data=df_reg
)

res = mod2.fit()
print(res.summary())
```

```python
# Model 3: Add interaction between Education and iden_pol_2
mod3 = smf.logit(
    formula='''
    accuracy ~ C(Q)
    + C(sexo)
    + C(grupo_edad)
    + C(zona_u_r)
    + C(Indigenous)
    + C(Education)
    + C(gse)
    + C(iden_pol_2)
    + C(religion_82)
    + C(Education):C(iden_pol_2)
    -1
    ''',
    data=df_reg
)

res = mod3.fit()
print(res.summary())
```

```python
# Model 4: Add interaction between religion_82 and iden_pol_2
mod4 = smf.logit(
    formula='''
    accuracy ~ C(Q)
    + C(sexo)
    + C(grupo_edad)
    + C(zona_u_r)
    + C(Indigenous)
    + C(Education)
    + C(gse)
    + C(iden_pol_2)
    + C(religion_82)
    + C(religion_82):C(iden_pol_2)
    -1
    ''',
    data=df_reg
)

res = mod4.fit()
print(res.summary())
```

## USA

```python
df_reg_usa = pd.DataFrame()

df = get_reg_df_eeuu(llama_1_eeuu_csv, 'elec_pres_144_a')
df['Q'] = 'presidencial'
df['Country'] = "EEUU"
df_reg_usa = pd.concat([df_reg_usa, df], ignore_index=True)

df = get_reg_df_eeuu(llama_2_eeuu_csv, 'religion_14')
df['Q'] = 'aborto'
df['Country'] = "EEUU"
df_reg_usa = pd.concat([df_reg_usa, df], ignore_index=True)

df_reg_usa
```

```python

df_reg_usa['Country'] = pd.Categorical(df_reg_usa['Country'], categories=['EEUU','Chile'], ordered=False)
df_reg_usa['sexo'] = pd.Categorical(df_reg_usa['sexo'], categories=['Hombre','Mujer'], ordered=False)
df_reg_usa['zona_u_r'] = pd.Categorical(df_reg_usa['zona_u_r'], categories=['Urbano','Rural'], ordered=False)
# df_reg['Indigenous'] = pd.Categorical(df_reg['Indigenous'], categories=['Non-indigenous','Indigenous'], ordered=False)
df_reg_usa['race'] = pd.Categorical(df_reg_usa['race'], categories=['White','Non-White'], ordered=False)

df_reg_usa['grupo_edad'] = pd.Categorical(df_reg_usa['grupo_edad'], categories=['joven','adulto','mayor'], ordered=False)
df_reg_usa['iden_pol_2'] = pd.Categorical(df_reg_usa['iden_pol_2'], categories=['Ninguna ideología','Izquierda','Centro','Derecha',], ordered=False)


```

```python
# Base Model
mod1 = smf.logit(
    formula='''
    accuracy ~ C(Q)
    + C(sexo)
    + C(grupo_edad)
    + C(zona_u_r)
    + C(race)
    + C(Education)
    + C(gse)
    + C(iden_pol_2)
    + C(religion_82)
    -1
    ''',
    data=df_reg_usa
)
res1 = mod1.fit()
print(res1.summary())
```

```python
#Model 2 (adds gender × ideology interaction)
mod2 = smf.logit(
    formula='''
    accuracy ~ C(Q)
    + C(sexo)
    + C(grupo_edad)
    + C(zona_u_r)
    + C(race)
    + C(Education)
    + C(gse)
    + C(iden_pol_2)
    + C(religion_82)
    + C(sexo):C(iden_pol_2)
    -1
    ''',
    data=df_reg_usa
)
res2 = mod2.fit()
print(res2.summary())
```

```python
# Model 3 (adds race × gender interaction)
mod3 = smf.logit(
    formula='''
    accuracy ~ C(Q)
    + C(sexo)
    + C(grupo_edad)
    + C(zona_u_r)
    + C(race)
    + C(Education)
    + C(gse)
    + C(iden_pol_2)
    + C(religion_82)
    + C(race):C(sexo)
    -1
    ''',
    data=df_reg_usa
)
res3 = mod3.fit()
print(res3.summary())
```
